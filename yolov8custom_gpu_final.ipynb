```python
import torch
import cv2
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import time
from glob import glob

class HelmetDataset(Dataset):
    def __init__(self, img_list, txt_list, transform=None):
        self.img_list = img_list
        self.txt_list = txt_list
        self.transform = transform

    def __len__(self):
        return len(self.img_list)

    def __getitem__(self, idx):
        img_path = self.img_list[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        if self.transform:
            img = self.transform(img)

        return img

# Transforms
transform = transforms.Compose([
    transforms.ToPILImage(),  # OpenCV로 로드된 이미지를 PIL 이미지로 변환
    transforms.Resize((256, 256)),  # 이미지 크기 조정
    transforms.ToTensor(),  # 텐서로 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # 정규화
])

# 데이터셋 및 데이터 로더 설정
# test_img_list = glob('/home/piai/project_yolo/ultralytics/poskick_helmet_7/test/images/*.jpg')
test_dataset = HelmetDataset(test_img_list, [], transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# # 모델 정의
model = YOLO("/home/piai/project_yolo/ultralytics/runs/detect/train25/weights/helmet_final.pt")
# model = model.to('cuda' if torch.cuda.is_available() else 'cpu')

# FPS 측정
start_time = time.time()
num_frames = 0

model.eval()
with torch.no_grad():
    for images in test_loader:
        images = images.to('cuda' if torch.cuda.is_available() else 'cpu')
        start = time.time()
        _ = model(images)  # 모델 예측 (실제 모델 사용시 변경)
        end = time.time()
        num_frames += images.size(0)
        print(f"Processed {images.size(0)} frames in {end - start:.3f} seconds.")

total_time = time.time() - start_time
fps = num_frames / total_time
print(f"Total time: {total_time:.2f} seconds, FPS: {fps:.2f}")
```

## 경량화

```python
import torch, torch.nn as nn
from torch.nn.utils import prune
from ultralytics import YOLO

model = YOLO("/home/piai/project_yolo/ultralytics/runs/detect/train25/weights/helmet_final.pt")

def sparsity(model):
    # 모델의 전체 sparsity 계산
    total_elements, zero_elements = 0, 0
    for p in model.parameters():
        total_elements += p.numel()
        zero_elements += (p == 0).sum().item()
    return zero_elements / total_elements

pruning_param = 0.1  # 점진적인 pruning을 위해 비율 감소

# 프루닝된 가중치를 영구적으로 제거
for name, module in model.model.named_modules():
    if isinstance(module, nn.Conv2d) or isinstance(module, nn.Linear):
        if hasattr(module, 'weight_mask'):
            prune.remove(module, 'weight')
```

```python
# Sparsity 확인
print(f'Model pruned to {sparsity(model.model):.3g} global sparsity')
```

```python
def measure_fps(model, val_loader, device):
    start_time = time.time()
    num_frames = 0
    model.eval()
    with torch.no_grad():
        for images in val_loader:
            images = images.to(device)
            start = time.time()
            predictions = model(images)
            end = time.time()
            num_frames += 1
            print(f"Frame {num_frames}: {end - start:.3f} seconds")

    total_time = time.time() - start_time
    print(f"Average inference time: {total_time / num_frames:.3f} seconds")
    print(f"FPS: {num_frames / total_time:.2f}")

# 예를 들어 모델 평가 후 FPS 측정
measure_fps(model, test_loader, device)
```
